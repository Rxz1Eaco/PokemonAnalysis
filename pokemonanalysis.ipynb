### Carregando os dados em xlsx

### Método Head 
- Tem como objetivo : Mostrar os primeiros dados do dataframe
- Exemplo: df_nome.head() ou df_nome.head(5) -> Aqui mostrar os 5° primeiros dados do dataframe

```python
df_pokemon.head(3)
```

### Método Tail 
- Tem como objetivo : Mostrar últimos dados do dataframe
- Exemplo: df_nome.tail() ou df_nome.tail(5) -> Aqui mostra os últimos 5° dados do dataframe

```python
df_pokemon.tail(3)
```

### Método Sample
- Tem como objetivo : Mostrar dados misturado 
- Exemplo: df_nome.sample() ou df_nome.sample(5) -> Aqui mostra os 5° dados misturados do dataframe

```python
df_pokemon.sample(3)
```

## Área de Estudo : Grama 🌱
### Pegando todos os pokémons com o primeiro tipo "grama" = 🌱

```python
grass_df_pokemon = df_pokemon[df_pokemon["primary_type"]=="grass"]
display(grass_df_pokemon)
```

```python
grass_water_df_pokemon = df_pokemon[(df_pokemon["primary_type"] == "grass") & (df_pokemon["secondary_type"] == "water")]
display(grass_water_df_pokemon)
```

## Pegando os pokémons do tipo Grama =🌱 e Venenoso = ☣️

```python
grass_poison_df_pokemon = df_pokemon[(df_pokemon["primary_type"]=="grass") & (df_pokemon["secondary_type"]=="poison")]
display(grass_poison_df_pokemon)
```
