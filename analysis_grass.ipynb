```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_pokemon = pd.read_excel("pokemon.xlsx")
df_pokemon.head(3)
```

```python
df_pokemon.info()
```

## Área de Estudo: Pokémon Grama 
### Pegando os pokémons com a tipagem Grama

```python
primary_type_pokemons_grama = df_pokemon[df_pokemon["primary_type"]=="grass"]
display(primary_type_pokemons_grama)
```

## Pegando os Pokémon com o Segundo Tipo = "Grama"

```python
secondary_type_pokemons_grama = df_pokemon[df_pokemon["secondary_type"]=="grass"]
display(secondary_type_pokemons_grama)
```

### Total De Pokémons de Grama

```python
primary_type_linhas,primary_type_colunas = primary_type_pokemons_grama.shape
secondary_type_linhas,secondary_type_colunas = secondary_type_pokemons_grama.shape
print(f"Existem um total de {primary_type_linhas} Pokémons do tipo Grama com a Primeira Tipagem Grama")
print(f"Existem um total de {secondary_type_linhas} Pokémons do tipo Grama com a Segunda Tipagem Grama")
print(f"Existe um total de {primary_type_linhas+secondary_type_linhas} Pokémons do tipo Grama(Tanto no primeiro tipo, quando no segundo tipo)")
```

### Melhores Pokémons do tipo Grama da Primeira e Segunda tipagem em relação ao Attack

```python
best_ataque_pokemons_grama = primary_type_pokemons_grama[primary_type_pokemons_grama['attack'] > 10]
best_ataque_pokemons_grama[["english_name","primary_type","attack"]]
```

```python
best_ataque_pokemons_grama = primary_type_pokemons_grama[primary_type_pokemons_grama['attack'] >= 0]
best_ataque_pokemons_grama_sorted = best_ataque_pokemons_grama.sort_values(by='attack', ascending=True)
display(best_ataque_pokemons_grama_sorted[["english_name", "primary_type", "attack"]])
```

```python
best_ataque_pokemons_grama = secondary_type_pokemons_grama[secondary_type_pokemons_grama['attack'] >= 0]
best_ataque_pokemons_grama_sorted = best_ataque_pokemons_grama.sort_values(by='attack', ascending=False)
display(best_ataque_pokemons_grama_sorted[["english_name", "secondary_type","attack"]])
```
