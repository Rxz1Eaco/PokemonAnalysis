```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_pokemon = pd.read_excel("pokemon.xlsx")
df_pokemon.head(3)
```

```python
df_pokemon.info()
```

### Contagem Totel por Geração

```python
contagem_tipos = df_pokemon["gen"].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(contagem_tipos.index, contagem_tipos, color=['#78C850', '#F08030', '#6890F0', '#F8D030', '#EE99AC'])
plt.title('Histograma de Pokémons por Geração')
plt.xlabel('Geraçaõ')
plt.ylabel('Quantidade')
plt.xticks(rotation=45, ha='right') 
plt.show()
```
