```python
import pandas as pd
```

```python
df_pokemon = pd.read_excel("pokemon.xlsx")
df_pokemon.sample(2)
```

Explicação sobre a Tabela:

- 0: Sem dano (o Pokémon é imune a ataques do tipo Normal).
- 1: Dano normal (nenhum aumento ou redução).
- 2: O dobro do dano (o Pokémon é fraco contra ataques do tipo Normal).
- 3: Triplo do dano.
- 4: Quadruplo do dano.

### Pokémons Fracos em Geral

```python
weakness_geral_df_pokemon = df_pokemon[(df_pokemon['against_normal' ] >= 2) 
                                       & (df_pokemon['against_fire' ] >= 2)
                                       & (df_pokemon['against_water' ] >= 2)
                                       & (df_pokemon['against_electric' ] <= 2)
                                       & (df_pokemon['against_grass' ] <= 2)
                                       & (df_pokemon['against_ice' ] <= 2)
                                       & (df_pokemon['against_fighting' ] <= 2)
                                       & (df_pokemon['against_poison' ] <= 2)
                                       & (df_pokemon['against_ground' ] <= 2)
                                       & (df_pokemon['against_flying' ] <= 2)
                                       & (df_pokemon['against_psychic' ] <= 2)
                                       & (df_pokemon['against_bug' ] <= 2)
                                       & (df_pokemon['against_rock' ] <= 2)
                                       & (df_pokemon['against_ghost' ] <= 2)
                                       & (df_pokemon['against_dragon' ] <= 2)
                                       & (df_pokemon['against_dark' ] <= 2)
                                       & (df_pokemon['against_steel' ] <= 2)
                                       & (df_pokemon['against_fairy' ] <= 2)]

weakness_geral_df_pokemon[["english_name","primary_type","against_normal","against_fire","against_water","against_electric","against_grass","against_ice","against_fighting","against_poison","against_ground","against_flying","against_psychic","against_bug","against_rock","against_ghost","against_dragon","against_dark","against_steel","against_fairy"]]
```

### Pokémons Fracos Contra o Tipo Normal 

```python

weakest_pokemon_top10 = df_pokemon["against_normal"].nlargest(4)

for idx in weakest_pokemon_top10.index:
    print("Nome do Pokémon:", df_pokemon.loc[idx, "english_name"])
    print("Tipos do Pokémon:", df_pokemon.loc[idx, "primary_type"])
    print("Tipos do Pokémon:", df_pokemon.loc[idx, "secondary_type"])
    print("Valor em against_normal:", df_pokemon.loc[idx, "against_normal"])
    print("\n")
```

```python
fraquezas_normal_vs_grama = df_pokemon[(df_pokemon['against_water' ] > 2) & (df_pokemon['against_ground' ] > 2)]

fraquezas_normal_vs_grama_sorted = fraquezas_normal_vs_grama.sort_values(by='against_normal', ascending=False)
display(fraquezas_normal_vs_grama_sorted[["english_name", "primary_type", "secondary_type","against_normal"]])
```
